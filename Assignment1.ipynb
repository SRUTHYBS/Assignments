# Qn5. Visualizations using the dataset.

```python
#Comparison of sepal length of each flower
#To import matplot and seaborn
import matplotlib.pyplot as plt
import seaborn as sns
#To plot the graph showing the values of sepal lengths of each flower
sns.violinplot(x='Classification',y='SL',data=data)
#To give title for figure
plt.title('Comparison of Sepal Length of each flower',fontsize=20)
```

```python
#Comparison of Sepal width of each flower
sns.boxplot(x='Classification',y='SW',data=data)
plt.title('Comparison of Sepal Width of each flower',fontsize=20)
```

```python
#Comparison of Petal lengths of each flower
sns.stripplot(x='PL',y='Classification',data=data)
plt.title('Comparison of Petal length of each flower',fontsize=20)
#T change the size of the figure
sns.set(rc= {"figure.figsize": (10,6)})
```

```python
#Comparison of Petal width of each flower
sns.stripplot(x='Classification',y='PW',data=data)
plt.title('Comparison of Petal width of each flower',fontsize=20)
sns.set(rc= {"figure.figsize": (10,6)})
```

```python
#Comparison of petal width and length of each flower
sns.stripplot(x='PL',y='PW',hue='Classification',data=data)
sns.set(rc= {"figure.figsize": (10,6)})
plt.title('Comparison of Petal features of each flower',fontsize=20)
```

```python
#Comparison of Sepal Width and length of each flower
sns.stripplot(x='SL',y='SW',hue='Classification',data=data)
sns.set(rc= {"figure.figsize": (10,6)})
plt.title('Comparison of Sepal features of each flower',fontsize=20)
```

```python
#Comparison of Sepal length and Petal length of each flower
sns.stripplot(x='PL',y='SL',hue='Classification',data=data)
sns.set(rc= {"figure.figsize": (10,6)})
plt.title('Comparison of Sepal length and Petal length of each flower',fontsize=20)
```

```python
#Comparison of Sepal length and Petal width of each flower
sns.stripplot(x='SL',y='PW',hue='Classification',data=data)
sns.set(rc= {"figure.figsize": (10,6)})
plt.title('Comparison of Sepal length and Petal width of each flower',fontsize=20)
```
